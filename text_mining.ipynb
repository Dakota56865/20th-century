```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from collections import Counter
from textblob import TextBlob
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
# Example assuming the text is stored in a file called 'twentieth_century.txt'
with open('key_events_20th_century.txt', 'r') as file:
    text_data = file.read()
```

```python
# Tokenizing the text
tokens = word_tokenize(text_data)

# Count word frequency
word_counts = Counter(tokens)

# Get the 10 most common words
most_common_words = word_counts.most_common(10)

# Plotting
words, counts = zip(*most_common_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.title('Top 10 Most Common Words')
plt.show()
```

```python
# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
tokens = [word.lower() for word in tokens if word.lower() not in stop_words and word not in string.punctuation]

# Recount word frequency
word_counts_clean = Counter(tokens)
most_common_words_clean = word_counts_clean.most_common(10)

# Plot again
words_clean, counts_clean = zip(*most_common_words_clean)
plt.figure(figsize=(10, 6))
plt.bar(words_clean, counts_clean)
plt.title('Top 10 Most Common Words After Removing Stopwords and Punctuation')
plt.show()
```

```python
blob = TextBlob(text_data)
tags = blob.tags
```

```python
# Count POS tags
pos_tags = [tag[1] for tag in tags]
pos_counts = Counter(pos_tags)

# Get top 10 POS tags
most_common_pos = pos_counts.most_common(10)

# Plot using seaborn
pos_labels, pos_freqs = zip(*most_common_pos)
plt.figure(figsize=(10, 6))
sns.barplot(x=list(pos_labels), y=list(pos_freqs))
plt.title('Top 10 POS Tags')
plt.show()
```

```python
# Filter tokens by POS
nouns = [word for word, pos in tags if pos.startswith('NN')]
verbs = [word for word, pos in tags if pos.startswith('VB')]
adjectives = [word for word, pos in tags if pos.startswith('JJ')]

# Count frequencies
nouns_count = Counter(nouns).most_common(15)
verbs_count = Counter(verbs).most_common(15)
adjectives_count = Counter(adjectives).most_common(15)

# Plot nouns
n_words, n_counts = zip(*nouns_count)
plt.figure(figsize=(10, 6))
sns.barplot(x=list(n_words), y=list(n_counts))
plt.title('Top 15 Nouns')
plt.show()

# Plot verbs
v_words, v_counts = zip(*verbs_count)
plt.figure(figsize=(10, 6))
sns.barplot(x=list(v_words), y=list(v_counts))
plt.title('Top 15 Verbs')
plt.show()

# Plot adjectives
a_words, a_counts = zip(*adjectives_count)
plt.figure(figsize=(10, 6))
sns.barplot(x=list(a_words), y=list(a_counts))
plt.title('Top 15 Adjectives')
plt.show()
```

```python
# List of countries (you can use more or a different list depending on your dataset)
countries = ['United States', 'Germany', 'France', 'China', 'Japan', 'Russia', 'Brazil', 'India']

# Create a dictionary of country mentions
country_counts = {country: text_data.lower().count(country.lower()) for country in countries}

# Create a DataFrame
df_countries = pd.DataFrame(list(country_counts.items()), columns=['Country', 'Mentions'])

# Plot the mentions
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Mentions', data=df_countries)
plt.title('Countries Mentioned in the Text')
plt.show()
```

```python
# Sentiment analysis
blob = TextBlob(text_data)
sentiment = blob.sentiment

# Plot sentiment scores
plt.figure(figsize=(6, 4))
plt.bar(['Polarity', 'Subjectivity'], [sentiment.polarity, sentiment.subjectivity])
plt.title('Sentiment Analysis')
plt.show()
```
